```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the aesthetic style for the plots
sns.set(style="whitegrid")
```

```python
# Load Titanic dataset from a URL 
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic_df = pd.read_csv(url)

# Display the first few rows
titanic_df.head()
```

```python
# Get basic information about the dataset
titanic_df.info()

# Get statistical summary of numerical features
titanic_df.describe()

# Check for missing values
titanic_df.isnull().sum()
```

```python
# Fill missing 'Age' values with the median age
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)

# Fill missing 'Embarked' values with the mode (most common value)
titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)

# Drop 'Cabin' column due to a large number of missing values
titanic_df.drop('Cabin', axis=1, inplace=True)

# Check if any missing values are left
titanic_df.isnull().sum()
```

```python
# Univariate analysis of 'Survived'
sns.countplot(x='Survived', data=titanic_df)
plt.title('Survival Count')
plt.show()

# Distribution of 'Age'
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['Age'], kde=True, bins=30)
plt.title('Age Distribution')
plt.show()

# Categorical features like 'Pclass' and 'Sex'
sns.countplot(x='Pclass', data=titanic_df)
plt.title('Passenger Class Distribution')
plt.show()

sns.countplot(x='Sex', data=titanic_df)
plt.title('Gender Distribution')
plt.show()
```

```python
# Survival rate by gender
sns.barplot(x='Sex', y='Survived', data=titanic_df)
plt.title('Survival Rate by Gender')
plt.show()

# Survival rate by passenger class
sns.barplot(x='Pclass', y='Survived', data=titanic_df)
plt.title('Survival Rate by Passenger Class')
plt.show()

# Age vs Survival
plt.figure(figsize=(10, 6))
sns.histplot(x='Age', hue='Survived', data=titanic_df, kde=True, bins=30)
plt.title('Age Distribution by Survival')
plt.show()
```

```python
# Pairplot for some important features
sns.pairplot(titanic_df[['Survived', 'Pclass', 'Sex', 'Age', 'Fare']], hue='Survived')
plt.show()

# Heatmap to visualize correlations
plt.figure(figsize=(12, 8))
sns.heatmap(titanic_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
